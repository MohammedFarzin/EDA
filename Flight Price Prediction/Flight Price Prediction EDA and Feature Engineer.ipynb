# Flight Price Prediction (EDA and Feature Engineering)

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('Data_train.xlsx')
df_train.head()
```

```python
df_test = pd.read_excel('Test_set.xlsx')
df_test.head()
```

```python
df_train.columns
```

```python
# appending df_train and df_test
df = df_train.append(df_test)
df.tail()
```

```python
df.info()
```

```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
# Date_of_Journey is object so we need to convert into integer
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df['Date'] = df['Date_of_Journey'].apply(lambda x: x.split('/')[0])
df['Month'] = df['Date_of_Journey'].apply(lambda x: x.split('/')[1])
df['Year'] = df['Date_of_Journey'].apply(lambda x: x.split('/')[2])
df.head()
```

```python
# Convert the object into integer
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
# Arrival time feature object into integer
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x: x.split(':')[0])
df['Arrival_Minute'] = df['Arrival_Time'].apply(lambda x: x.split(':')[1])
df.head()
```

```python
# Convert object into integer
df[['Arrival_Hour', 'Arrival_Minute']] = df[['Arrival_Hour', 'Arrival_Minute']].astype(int)

# Drop the Arrival_Time Feature
df.drop('Arrival_Time', axis=1, inplace=True)
df.info()
```

```python
# Departure type converting from object to integer
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x: x.split(':')[0])
df['Dep_Minute'] = df['Dep_Time'].apply(lambda x: x.split(':')[1])

# Coverting object into integer
df[['Dep_Hour', 'Dep_Minute']] = df[['Dep_Hour', 'Dep_Minute']].astype(int) 

# Dropping the Dep_Time Feature
df.drop('Dep_Time', axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
# Total_Stops Feature converting object into integer
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan': 1})
```

```python
df.drop('Route', axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
# Duration
df['Duration_Hour'] = df['Duration'].apply(lambda x: x.split(' ')[0].split('h')[0])
df.head()
```

```python
df[df['Duration_Hour'] == '5m']
```

```python
df.drop([6474, 2660], axis=0, inplace=True)
```

```python
df['Duration_Hour'] = df['Duration_Hour'].astype(int)
df.info()
```

```python
df['Duration'].apply(lambda x: int(x.split()[0][:-1]))
```

```python
df['Duration_Minutes'] = df['Duration'].apply(lambda x: (int(x.split()[0][:-1]) * 60) + int(x.split()[1][:-1]) if len(x.split()) > 1 else int(x.split()[0][:-1]) * 60)
df.head()
```

```python
# Airline feature
df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder() 
df['Airline'] = labelencoder.fit_transform(df['Airline'])
df['Source'] = labelencoder.fit_transform(df['Source'])
df['Destination'] = labelencoder.fit_transform(df['Destination'])
df['Additional_Info'] = labelencoder.fit_transform(df['Additional_Info'])
```
