# Black Friday Dataset EDA and Feature Engineering

Cleaning and preparing the data for model training

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset
df_train = pd.read_csv('black_friday_train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('black_friday_test.csv')
df_test.head()
```

```python
# Merge both train and test data
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Handling categorical Gender
df['Gender'] = df['Gender'].map({'F':0 , 'M':1})
df.head()
```

```python
# Handling categorical Age
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
df.head()
```

```python
# Label encoding
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['Age'] = label_encoder.fit_transform(df['Age'])
df['Age'].unique()
```

```python
df.head()
```

```python
# Fixing categorical variable City_Category
df_city = pd.get_dummies(df['City_Category'], drop_first=True)
df_city.head()
```

```python
# concatnating df_city to df
df = pd.concat([df, df_city], axis=1)
df.head()
```

```python
# Drop City_Category Feature
df.drop('City_Category', axis=1, inplace=True)
df.head()
```

```python
# Missing values
# checking null values
df.isnull().sum()
```

```python
# checking the missing values
df['Product_Category_2'].isnull().sum()
```

```python
# mode value of product_category_2
df['Product_Category_2'].mode()[0]
```

```python
# Replacing the null values with mode
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
# Checking is there any null value in product_category_2
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].value_counts()
```

```python
# mode value of Product_Category_3
df['Product_Category_3'].mode()[0]
```

```python
# replacing null values with mode in Produt_Category_3
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
df['Product_Category_3'].isnull().sum()
```

```python
df.shape
```

```python
# Fixing categorical variable Stay_In_Current_City_Years
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+', '')
df.head()
```

```python
df.info()
```

```python
# convert object into integers
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype('int')
```

```python
df['B'] = df['B'].astype('int')
df['C'] = df['C'].astype('int')
```
