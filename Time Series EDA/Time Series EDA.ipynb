## Time Resampling

```python
df_tesla.head()
```

```python
# rule = 'A' means year end frequency
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
## Quarterly start frquency
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
# Buisness End Frequency
df_tesla.resample(rule='BA').max().plot()
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar', figsize=(19, 6))
```

```python
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
df_tesla.head(31)
```

```python
 df_tesla[['Open', 'Open: 30 days rolling']].plot(figsize=(12, 5))
```
